```python
df=df.sort_values(by='Make', ascending=True) #Gives sorted value based on the col provided
df.head()
```

```python
#Notice that the index is now unordered. we can reset index
df.reset_index(inplace=True, drop=True) #drop=True deletes previous indices
df.head()
```

```python
df.head()
```

```python
df.dtypes #gives the datatypes of each columns
```

```python
df.T #Transpose DataFrame
```

```python
#Selecting only required columns
cols = ['YEAR', 'Make', 'Model', '(kW)', '(km)']
some_cols = df[cols]
```

```python
some_cols.head()
```

```python
#Selecting only required rows
df.iloc[4:7,]
```

```python
#Selecting only the required rows and columns
df.iloc[15:20, [1,7,9]]
```

```python
df.head()
```

```python
df.nunique()  #Gives number of unique values for  each of the rows
```

```python
df['YEAR'].unique()  #unique functin gives the unique values for a given column name
```

```python
df['YEAR'].nunique()
```

```python
#We can even perform conditional selection
df[df.YEAR==2015] #Gives the data for the year 2015 only
```

```python
#Multipli conditions are also accepted; however, each comparison expression should be included inside paranthesis
df[(df.YEAR==2015) & ((df.Make=='SMART') | (df.Make=='TESLA'))]
```

```python
#New columns can be added based on the values of previous ones
df['Avg (kWh/100 km)'] = (df['CITY (kWh/100 km)'] + df['HWY (kWh/100 km)'] + df['COMB (kWh/100 km)'])/3
```

```python
df[['YEAR', 'Make', 'Model','Avg (kWh/100 km)']].head()
```

```python
#we can declare our column as index
df.set_index('YEAR', drop=True, inplace=True)
df.head()
```

```python
#We can go back to the previous auto generated index using reset_index
df.reset_index(inplace=True, drop=False) #drop=True will remove the previous index i.e. YEAR in this case
df.head()
```

```python
#We can even set multiple indices called MultiIndexing
df.set_index(['YEAR', 'Make', 'Model'], drop=True, inplace=True)
df = df.sort_values(by='YEAR')
```

```python
df.head(25) #Notice the hierarchical sturcture of YEAR, Make, and Model values in the output below
```
